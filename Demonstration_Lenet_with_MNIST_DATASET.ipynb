```python
from tensorflow import keras
from keras.datasets import mnist
# ab keras ke layer dalne hai 
from tensorflow.keras.layers import Conv2D,MaxPooling2D,AveragePooling2D

# ANN ke liye DENSE layer niklne hai 

from keras.layers import Dense,Flatten
from keras.models import Sequential
```

```python
# ab data set ko load krna hai 

(x_train,y_train),(x_test,y_test)=mnist.load_data()
```

```python
x_train.shape    # (60000, 28, 28) eska mtlb kya hai []
```

# convalution 2D  me jarurat hota hai >> batch_size(datako kitne size me bate ho),width,height,channels[kis color ke liye [grayscale=1,] ab hame data ka size change krna hoga heigth,width ke form me ]

```python
# ab esko krenge kaise 

x_train=x_train.reshape(x_train.shape[0],28,28,1)  # ab same kam app x_test ke liye bhi kro 

x_test=x_test.reshape(x_test.shape[0],28,28,1)


x_train.shape
```

```python
x_test.shape
```

```python
# ab Normalized 

x_train=x_train/255  # esliye kiye hai 255 se divide 
x_test=x_test/255   # kyu ki pixel 256 ka hota hai esaliye 255 tak jata hai [ formula:- x-Xmin/x-Xmin]
```

```python
y_train  # y_train ko dekhiye dhyan se [ye ak tarah se categorical data hai [5 is the label ]]
```

```python
# onehot Encoding  keras me kaise krenge 

y_train=keras.utils.to_categorical(y_train,10) # kisko krna hai than kitne part me krna hai ye dalna parta hai 

y_test=keras.utils.to_categorical(y_test,10)
```

# ab apko model artechture  create krna hai 

# lenet me 7 layer the  to lenet ka 7 layer banao 

```python
model=Sequential()# model create ho gya 

# ab model me layer  add kro 

model.add(Conv2D(6, kernel_size=(5,5), activation="relu", input_shape=(28,28,1)))  # kawal phle layer me he input_shape dena jaruri hai 
model.add(MaxPooling2D(pool_size=(2,2)))

 # same kam ko repeate kr dete hai 
model.add(Conv2D(16, kernel_size=(5,5), activation="relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(120, kernel_size=(5,5), activation="relu",padding="same"))
#model.add(MaxPooling2D(pool_size=(2,2)))  # yaha pr max pooling dena haide sakte ho nhi dena hai mt do 
model.add(Flatten())
model.add(Dense(84,activation="relu"))  # dense layer multiple bhi de sakte ho 

model.add(Dense(10,activation="softmax"))  # output layer kawal 10 hoga 

 # model tiyar hai ab esko summary kr ke dekhte hai 

model.summary()
```

```python
# ab model ko compile krenge 

model.compile(loss=keras.metrics.categorical_crossentropy,optimizer=keras.optimizers.Adam(),metrics=["accuracy"])
model.fit(x_train,y_train,batch_size=128,epochs=20,verbose=1,validation_data=(x_test,y_test))
```

# dekh sakte hai model train ho gya 

```python
score=model.evaluate(x_test,y_test)
print("test loss is =  ",score[0])
print("test accuracy is ",score[1])
```

#   yaha se pytorch implementation krenge  mnist datasets for lenet

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class LeNet(nn.Module):
    def __init__(self):
        super(LeNet, self).__init__()
        self.conv1 = nn.Conv2d(1, 6, kernel_size=5)  # 6 filter hai 5 by 5 ka 
        self.pool = nn.MaxPool2d(2, 2)  # 2 by 2 ka pooling hai 
        self.conv2 = nn.Conv2d(6, 16, kernel_size=5) # 6 filter 16 by 16 ka 
        self.conv3 = nn.Conv2d(16, 120, kernel_size=5, padding=2)  # 16 filter 120 by 120 ka 
        self.fc1 = nn.Linear(120*4*4, 84)   # yaha nn.Linear hoga and 84 neuron me 
        self.fc2 = nn.Linear(84, 10) # 84 neuron me 10 output layer aiya hai 

    def forward(self, x):
        x = F.relu(self.conv1(x))  # 24x24x6
        x = self.pool(x)           # 12x12x6
        x = F.relu(self.conv2(x))  # 8x8x16
        x = self.pool(x)           # 4x4x16
        x = F.relu(self.conv3(x))  # 4x4x120
        x = x.view(-1, 120*4*4)  # flatten kr rhe hai yaha pr 
        x = F.relu(self.fc1(x))  # dense layer create kiye hai 
        x = self.fc2(x) # output hai ye 
        return x

# model banake test karte hain
model = LeNet()
print(model)


import torch.optim as optim
criterion=nn.CrossEntropyLoss()
optimizer=optim.Adam(model.parameters(),lr=0.001)
```

```python
# move model to GPU if available
from torch.utils.data import TensorDataset, DataLoader
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = model.to(device)

# convert your existing numpy data to PyTorch tensors
x_train_tensor = torch.tensor(x_train, dtype=torch.float32).permute(0, 3, 1, 2)
y_train_tensor = torch.tensor(y_train.argmax(axis=1), dtype=torch.long)

x_test_tensor = torch.tensor(x_test, dtype=torch.float32).permute(0, 3, 1, 2)
y_test_tensor = torch.tensor(y_test.argmax(axis=1), dtype=torch.long)

# create datasets and loaders
batch_size = 128

train_dataset = TensorDataset(x_train_tensor, y_train_tensor)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

test_dataset = TensorDataset(x_test_tensor, y_test_tensor)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

# define loss and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# train the model
epochs = 10
for epoch in range(epochs):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

    print("Epoch {}/{} | Loss: {:.4f} | Train Accuracy: {:.2f}%".format(
        epoch + 1, epochs, running_loss / len(train_loader), 100 * correct / total))

# evaluate on test data
model.eval()
correct = 0
total = 0

with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print("Test Accuracy: {:.2f}%".format(100 * correct / total))
```
